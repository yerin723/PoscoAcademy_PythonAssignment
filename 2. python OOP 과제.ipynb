# 0503 클래스 실습 예제
교육생 본인의 인스턴스를 생성하시오.
- 변수 : 성명, 나이, 키, 몸무게
- 메서드 : 나이 1증가, 키 변화, 몸무게 변화

```python
class Person:
    def __init__(self, name, age, h, w):
        self.name = name
        self.age = age
        self.h = h
        self.w = w
    
    def print_info(self):
        print('이름:', self.name)
        print('나이:', self.age)
        print('키:', self.h)
        print('몸무게:', self.w)
        print()
        
    def add_age(self):
        self.age += 1
    
    def change_h(self, new_h):
        self.h = new_h
        
    def change_w(self, new_w):
        self.w = new_w
        
a = Person('김예린', 28, 158, 50)
a.print_info()

a.change_h(180)
a.change_w(77)
a.print_info()
```

# 0503 클래스 실습 문제 1 - 은행 계좌 클래스 만들기
- 계좌 클래스의 __init__인자는 balance와 name을 지님
- 계좌 생성시 초기값으로 balance와 name을 지정 가능
- 지정 안할 시 각 초기값은 0과 "none"을 가질 것 
- 계좌의 기능은 입금, 출금, 계좌정보출력
- 출금은 출금하고자하는 금액이 잔고에 있을 때 출금할 것

- 실행 시나리오
1) 계좌에 400원 입금 후 600원 출금하여 "잔액 부족" 출력하기
2) 계좌에 1000원 입금 후 900원 출금하여 잔고 100으로 만들기

```python
class BankAccount:
    def __init__(self, balance=0, name='none'):
        self.balance = balance
        self.name = name
        
    def doposit(self, amount):
        self.balance += amount
        print('입금 금액: ', self.balance)
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print('출금 금액: ', self.balance)
        else:
            print(amount, '원 출금실패')
            print('이유 : 잔액부족')
    
    def get_info(self):
        print('이름: ', self.name)
        print('계좌 잔고: ', self.balance)
        print()
        
a = BankAccount(100, "Yerin Kim")
a.get_info()

a.doposit(400)
a.withdraw(600)
a.get_info() #계좌 정보 출력

b = BankAccount(0, "posco")
b.doposit(1000)
b.withdraw(900)
b.get_info() #계좌 정보 출력
```

# 0503 클래스 실습 문제 2 - 최소 잔액을 유지해야 하는 계좌 클래스 만들기
- 기본적으로 기존 실습 문제 1의 계좌와 기능 동일함
- 하지만 출금시 잔액이 최소 잔액 미만이 되면 출금 못함 그리고 "최소 잔액을 유지해야 합니다" 메시지 출력
- 힌트 : 상속 및 오버라이딩 활용

```python
class MinimumBalanceAccount(BankAccount):
    def __init__(self, balance=0, name='none', min_val=0):
        super().__init__(balance, name)
        self.min_val = min_val
        
    def withdraw(self, amount):
        if self.balance - amount < self.min_val:
            print('최소 잔액을 유지해야 합니다.')
        else:
            self.balance -= amount
    
c = MinimumBalanceAccount(500, "Yerin Kim", 1000) #기준값은 1000원
c.withdraw(1000)
c.get_info()
```

# 0503 클래스 과제 - Python의 기본 자료구조 set을 모방한 클래스 Set을 구현하세요.
- set : 임의의 원소를 중복 없이, 순서 없이 담는 집합형 자료구조

```python
class Set:
    def __init__(self, l=[]):
        #list를 받아 중복 제거
        self.l = []
        for i in l:
            if i not in self.l:
                self.l.append(i)
    
    #Set에 elem이 존재하지 않으면 추가
    def add(self, elem):
        if elem not in self.l:
            self.l.append(elem)
        else:
            print('exist')
    
    #Set에 elem이 존재하면 삭제
    def discard(self, elem):   
        if elem in self.l:
            self.l.remove(elem)
        else:
            print('not exist')
            return elem
    
    #Set에 존재하는 모든 원소 삭제
    def clear(self):
        self.l = []
    
    #Set에 존재하는 원소 개수 반환
    def __len__(self):
        count = 0
        for i in self.l:
            count += 1
        return count
    
    #Set에 존재하는 원소를 '{1, 2, 3}'의 형태로 반환
    def __str__(self): # print(self) = string / self = Set
        return "{" + ', '.join(map(str, self.l)) + "}"
    
    #in 멤버체크 - Set에 elem이 존재하면 참 반환, 아니면 거짓 반환
    def __contains__(self, elem):
        # in
        if elem in self.l:
            return True
        else:
            return False
            
    #self가 other의 부분집합이면 참 반환, 아니면 거짓 반환
    def __le__(self, other):
        # <=
        for i in self.l:
            if i in other.l:
                return True
            else:
                return False

    #other가 self의 부분집합이면 참 반환, 아니면 거짓 반환
    def __ge__(self, other):
        # >=
        result = True
        for i in other.l:
            if i not in self.l:
                return False
            else:
                result = True
        return result #모든 원소가 self에 존재하는 경우, flag 값이 변하지않기 때문에 True
        
    #self와 other의 원소를 모두 포함하는 합집합 Set 반환
    def __or__(self, other):
        result = Set()
        for i in self.l:
            result.l.append(i)            
        for i in other.l:
            if i not in self.l:
                result.l.append(i)
        return result
    
    #self와 other가 공통으로 포함하는 원소를 포함하는 교집합 Set 반환
    def __and__(self, other):
        result = Set()
        for i in self.l:
            if i in other.l:
                result.l.append(i)
        return result
           
    # self - other : self의 원소 중 other에 없는 원소만을 포함하는 차집합 Set 반환
    def __sub__(self, other):
        result = Set()
        for i in self.l:
            if i not in other.l:
                result.l.append(i)
        return result
    
    def __ior__(self, other):
        # |= : in-place 연산자(=복합 대입 연산자)
        result = self | other
        self.l = result.l.copy()
        return self
    
    def __iand__(self, other):
        # &=
        result = self & other
        self.l = result.l.copy()
        return self
    
    def __isub__(self, other):
        # -=
        result = self - other
        self.l = result.l.copy()
        return self

a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

a = Set([1,2,3])
b = Set([2,3,4])

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)

a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
