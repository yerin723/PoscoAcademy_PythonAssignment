```python
#데이터 읽기
import os
import sys
def getStuInfo(file_name="students.txt"): 
    lineslist = list()
    with open(file_name, "r" ) as f : #with 블록을 벗어는 순간 파일을 자동으로 닫음
        while True:    
            line = f.readline()
            linelist = line.strip().split('\t')
            if line == '':
                break
            lineslist.append(linelist)
    return lineslist
    
while True:
    #파일명 입력
    if len(sys.argv) > 1:
        file_name = sys.argv[1]
        students_info = getStuInfo(file_name)
        break
    else:
        file_name = input("Enter a file name: ")
        if os.path.exists(file_name) == True:
            students_info = getStuInfo(file_name)
            break  # 파일이 존재하는 경우 반복문 종료
        elif file_name == '':
            students_info = getStuInfo()
            break  # 파일명을 입력하지 않은 경우에도 반복문 종료
        else:
            pass

def stuUI(): # 헤더 출력문
    print('{:>10s}{:>15s}{:>10s}{:>8s}{:>10s}{:>8s}'.format('Student','Name','Midterm','Final','Average','Grade'))
    print('-'*65)
    
def getStuScore(students_info):
    for num in students_info:
        avg = (float(num[2]) + float(num[3])) / 2 
        grade = ''
        if avg >= 90: 
            grade = 'A'
        elif avg >= 80: 
            grade = 'B'
        elif avg >= 70: 
            grade = 'C'
        elif avg >= 60: 
            grade = 'D'
        else:
            grade = 'F'
        num.append(avg)
        num.append(grade)
    return students_info

def updateStuScore(student_info):
    avg = (int(student_info[2]) + int(student_info[3])) / 2 
    student_info[4] = avg 
    if avg >= 90: 
        student_info[5] = 'A'
    elif avg >= 80: 
        student_info[5] = 'B'
    elif avg >= 70: 
        student_info[5] = 'C'
    elif avg >= 60: 
        student_info[5] = 'D'
    else:
        student_info[5] = 'F'
    return student_info
           
def show_function(students_info):
    students_info = getStuScore(students_info)
    students_info = sorted(students_info, key=lambda x: x[4], reverse=True) # 평균 기준으로 내림차순 정렬
    stuUI() 
    for stu in students_info:
        print('{:>10s}{:>15s}{:>8d}{:>8d}{:>10.1f}{:>8s}'.format(stu[0],stu[1],int(stu[2]),int(stu[3]),float(stu[4]),stu[5]))
    
def search_function(students_info):
    students_info = getStuScore(students_info)
    students_info = sorted(students_info, key=lambda x: x[4], reverse=True) # 평균 기준으로 내림차순 정렬
    result = False
    search = input('Student ID: ')
    for stu in students_info: 
        if search in stu:
            stuUI()
            print('{:>10s}{:>15s}{:>8d}{:>8d}{:>10.1f}{:>8s}'.format(stu[0],stu[1],int(stu[2]),int(stu[3]),float(stu[4]),stu[5]))
            result = True
            break
    if not result: # result = False일 때 조건식이 참이되어서 print문 출력
        print('NO SUCH PERSON.')
    
def changescore_function(students_info):
    students_info = getStuScore(students_info)
    students_info = sorted(students_info, key=lambda x: x[4], reverse=True) # 평균 기준으로 내림차순 정렬
    result = False
    search = input('Student ID: ')
    for i, stu_list in enumerate(students_info):
        if search in stu_list:
            result = True
            change_value = input('Mid/Final? ')
            if change_value == 'mid' or change_value == 'final':
                change_score = int(input('Input new score: '))
                if  0 <= change_score <= 100:
                    if change_value == 'mid':
                        students_info[i][2] = change_score
                        students_info[i] = updateStuScore(students_info[i])
                    else:
                        students_info[i][3] = change_score
                        students_info[i] = updateStuScore(students_info[i])
                else:
                    pass # 0~100 외의 값이 입력된 경우 
            else:
                pass # mid/final 외의 값이 입력된 경우
    if not result:
        print('NO SUCH PERSON.')    
    return students_info
    
def add_function(students_info):
    search = input('Student ID: ')
    for stu_list in students_info: 
        if search in stu_list[0]:
            print('ALREADY EXISTS.')
            return students_info

    name = input('Name: ')
    mid = int(input('Midterm Score: '))
    final = int(input('Final Score: '))
    
    new_student = [search, name, mid, final]
    new_student = getStuScore([new_student])[0]
    students_info.append(new_student)
    
    print('Student added.')
    return students_info

def searchgrade_function(students_info):
    students_info = getStuScore(students_info)
    students_info = sorted(students_info, key=lambda x: x[4], reverse=True) # 평균 기준으로 내림차순 정렬
    search = input('Grade to search: ').upper()
    if search not in ['A', 'B', 'C', 'D', 'F']:
        pass
    else:
        li = []
        for stu_list in students_info: 
            if search in stu_list:
                li.append(stu_list)      
        if li == []:
            print('NO RESULTS.')
        else:
            stuUI()
            for stu in li:
                print('{:>10s}{:>15s}{:>8d}{:>8d}{:>10.1f}{:>8s}'.format(stu[0],stu[1],int(stu[2]),int(stu[3]),float(stu[4]),stu[5]))           
               
def REMOVE_function(students_info):
    if not students_info:
        print('List is empty.')
        return students_info
    
    search = input('Student ID: ')
    student_found = False
    for stu_list in students_info:
        if search in stu_list:
            students_info.remove(stu_list)
            print('Student removed.')
            student_found = True
    if not student_found:
        print('NO SUCH PERSON.')
    return students_info

def quit_function(students_info):
    students_info = getStuScore(students_info)
    students_info = sorted(students_info, key=lambda x: x[4], reverse=True) # 평균 기준으로 내림차순 정렬
    while True:
        question = input('Save data?[yes/no] ')
        if question == 'yes':
            file_name = input('File name: ')
            with open(file_name, "w") as final_file:
                for stu in students_info:
                    final_file.write('{:s}\t{:s}\t{:d}\t{:d}\n'.format(stu[0],stu[1],int(stu[2]),int(stu[3])))
            break       
        elif question == 'no':
            break
        else:
            pass

# 메인 프로그램
while True:
    command = input("# ").lower() # 명령어를 모두 소문자로 변환
    if command == "show":
        show_function(students_info)
    elif command == "search":
        search_function(students_info)
    elif command == "changescore":
        students_info = changescore_function(students_info)
    elif command == "add":
        students_info = add_function(students_info)
    elif command == "searchgrade":
        searchgrade_function(students_info)
    elif command == "remove":
        students_info = REMOVE_function(students_info)
    elif command == "quit":
        quit_function(students_info)
        break
    else:
        print("wrong input!")
```
